```python
# Time Series Data Visualization
#rolling statistics is used for real time data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

20

# Create time-series data
date_range = pd.date_range(start='2024-01-01', periods=30, freq='D')
values = np.random.randint(50, 100, size=30)

df = pd.DataFrame({'Date': date_range, 'Value': values})
df.set_index('Date', inplace=True)

# Plot time-series data
plt.figure()
plt.plot(df.index, df['Value'])
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Data")
plt.xticks(rotation=80)
plt.show()

# Rolling statistics
rolling_mean = df['Value'].rolling(window=5).mean()
rolling_std = df['Value'].rolling(window=5).std()

# Plot rolling statistics
plt.figure()
plt.plot(df['Value'], label='Original Data')
plt.plot(rolling_mean, label='Rolling Mean')
plt.plot(rolling_std, label='Rolling Std Dev')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Rolling Statistics of Time Series")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
