```python
# Geospatial Data Visualization

import pandas as pd
import folium
import plotly.express as px
```

```python
# Create sample geospatial data
data = {
'City': ['Mumbai', 'Delhi', 'Bengaluru', 'Chennai', 'Kolkata'],
'Latitude': [19.0760, 28.7041, 12.9716, 13.0827, 22.5726],
'Longitude': [72.8777, 77.1025, 77.5946, 80.2707, 88.3639]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Create Folium Map
map_india = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
print(map_india)
```

```python
# Add markers
for i in range(len(df)):
  folium.Marker(
      location=[df.loc[i, 'Latitude'], df.loc[i, 'Longitude']],
      popup=df.loc[i, 'City']
      ).add_to(map_india)
```

```python
# Plotly Scatter Geo Map
fig = px.scatter_geo(
df,
lat='Latitude',
lon='Longitude',
text='City',
title='Geospatial Visualization using Plotly'
)
fig.show()
```
